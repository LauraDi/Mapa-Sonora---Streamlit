```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('CensoSonora2020.csv', encoding='latin-1',index_col = 0)  
```

```python
from geopy import Nominatim

locator = Nominatim(user_agent="myGeocoder")
```

```python
df["lon"] = ""
df["lat"] = ""
df
```

```python
for i in df.index: 
    try:
        #direccion= df['DOMICILIO'][i]
        municipio= df['municipio'][i]

#        ubicacion= direccion + "," + municipio
        ubicacion= municipio + ", Sonora, Mexico"

        location = locator.geocode(ubicacion)
        df['lat'][i] = location.latitude
        df['lon'][i] = location.longitude
    except:
        print(ubicacion)
        
df
```

```python
df.to_csv("datos.csv")
```
